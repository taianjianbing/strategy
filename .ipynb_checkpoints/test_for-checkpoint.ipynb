```python
from gmsdk.api import StrategyBase
from gmsdk.api import md
from operator import truediv

import pandas as pd
import numpy as np
import datetime

'''
请在Strategy中修改个人账号密码和策略ID
'''


class Strategy(StrategyBase):
    
    def __init__(self, *args, **kwargs):
        super(Strategy, self).__init__(*args, **kwargs)
        self.yb=60
        self.stocks_pool = []
        self.stocks_pool_pd = pd.DataFrame()
        self.stocks_pool_pd2 = pd.DataFrame()
    def get_stocks_pool(self):
        instruments1 = self.get_instruments('SHSE', 1, 1)
        #print instruments1
        instruments2 = self.get_instruments('SZSE', 1, 1)
        symbol_list1 = set(instrument.symbol for instrument in instruments1+instruments2)
        constituents = self.get_constituents('SHSE.000300')
        symbol_list2 = set(constituent.symbol for constituent in constituents)
        symbol_list =  list(symbol_list1 & symbol_list2)
        self.stocks_pool = symbol_list
    def test(self):
        
        return self.get_financial_index('SHSE.600000', '2016-06-14', '2016-08-14')
    def FF(self, start_time, end_time):
        temp_dic = {}
        
        for stock in self.stocks_pool[:5]:
            stock_date = []
            stock_market_index = []
            stock_close = []
            stock_market_financial_index = []

            stock_dailybars = self.get_last_n_dailybars(stock, self.yb)
            
            for dailybar in stock_dailybars:
                daily_time = dailybar.strtime[:-6].replace("T", " ")
                stock_date.append(dailybar.strtime[:10])
                
                stock_close.append(dailybar.close)
                stock_market_financial_index.append(self.get_financial_index(stock, self.getDay(daily_time, -90),
                                                                                       daily_time)[0])
            
            stk = self.get_last_n_market_index(stock, self.yb)
            stock_market_value = [index.market_value for index in stk]
            stock_equity = [index.equity for index in stock_market_financial_index]

            stock_BTM = map(truediv, stock_equity, stock_market_value)
           
        
            stock_total_asset = [index.total_asset for index in stock_market_financial_index]
           
            temp_dic[stock] = {'Date': stock_date, "stock_market_value": stock_market_value,
                               'stock_equity':stock_equity, 'stock_BTM':stock_BTM,
                               'stock_total_asset':stock_total_asset, 'stock_close': stock_close}    
        
        self.stocks_pool_pd2 = pd.DataFrame(temp_dic)
        reform = {(outerKey, innerKey): values for outerKey, innerDict in temp_dic.iteritems() for innerKey, values in innerDict.iteritems()}

        self.stocks_pool_pd = pd.DataFrame(reform)
        self.stocks_pool_pd.to_csv('stocks_pool_dic.csv')
                                                    
    def getDay(self, present, dt):

        t = datetime.datetime.strptime(present, '%Y-%m-%d %H:%M:%S') + datetime.timedelta(days=dt)
        return datetime.datetime.strftime(t, '%Y-%m-%d %H:%M:%S')

strategy = Strategy()

mystrategy = Strategy(username='18660821753', # 请修改账号
        password='long199124', # 请修改密码
        strategy_id='9f74b6f0-516e-11e7-a854-5254006af232', # 请修改策略ID
        mode=2,
        td_addr='106.75.147.38:7001')
mystrategy.get_stocks_pool()

mystrategy.FF("2016-09-01 09:30:00", "2016-10-01 09:30:00")

print mystrategy.test()

```
